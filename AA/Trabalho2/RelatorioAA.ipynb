Este trabalho consiste na implemenção de diferentes algoritmos em Python para gerar diferentes modelos de AA, tendo estes modelos comportamentes diferentes face a um determinado conjunto de teste.

Conjunto de dados de treino que o algoritmo utiliza para criar o modelo:

```python
import pandas as pd
datrain = pd.read_csv('train.csv')
datrain.head(20)
```

## Preparação dos dados utilizados pelos Algoritmos:

```python
del datrain['Program']
del datrain['Id']
#Estas duas linhas têm como objetivo retirar do dataframe as colunas "Id" e "Program" visto que não achamos relevantes para a criação do modelo.
#Ou seja os algoritmos utilizam os restantes atributos na criação dos modelos.

x_train = datrain.drop('Failure', axis=1).values
y_train = datrain['Failure'].values
#Estas duas linhas preparam o x_train e o y_train para serem utilizados pelo algoritmo.
```

## Algoritmo Logistic Regression


```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', max_iter=500)
model.fit(x_train, y_train)
```

## Comportamento de cada modelo face ao determinado conjunto de dados:

```python
import pandas as pd
datest = pd.read_csv('test.csv')
datest.head(20)
```

```python
del datest['Program']
del datest['Id']
#Remover as colunas que nao achamos relevantes.
test = datest.values
#Preparar o data para ser utilizado no predict

predicted = model.predict(test)
#no predicted encontra-se os resultos previsto com base no modelo utilizado
```

```python
dfid= datest['Id'] #dataframe apenas com a coluna "Id"
dffail= pd.DataFrame({'Failure': predicted}) #um novo dataframe que tem os valores de "Failure
frames = (dfid, dffail) #os dois dataframes
datasubm = pd.concat(frames, axis=1) # com o concat() juntamos os 2 dataframes num só
datasubm.to_csv(r'SubmissionLogi.csv', index = False) #o Dataframe criado com as colunas "Id" e "Failure"
```
